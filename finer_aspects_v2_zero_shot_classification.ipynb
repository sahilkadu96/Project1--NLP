```python
from transformers import pipeline
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/Omdena/AI for soft skills/Finer_aspects_v2_summary.csv')
df
```

```python
classifier = pipeline("zero-shot-classification",
                      model="facebook/bart-large-mnli")
```

```python
def zero_shot_score(dataframe):
  zero_shot_scores = []
  for i in range(len(dataframe)):
    sequence_to_classify = dataframe['Summary'][i]
    candidate_labels = dataframe['Finer aspects'][i]
    result = classifier(sequence_to_classify, candidate_labels)
    zero_shot_scores.append(result['scores'][0])
  return zero_shot_scores

```

```python
scores = zero_shot_score(df)
```

```python
scores
```

```python
df['zero_shot_score'] = scores
```
