```python
from transformers import pipeline

import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('/content/drive/MyDrive/Omdena/AI for soft skills/Finer_aspect_v2.xlsx')
df
```

```python
# model : https://huggingface.co/facebook/bart-large-cnn
summarizer = pipeline('summarization', model="facebook/bart-large-cnn")
```

```python
def summarize_content(content):
  return summarizer(content)[0]['summary_text'].strip()

df['Summary'] = df['Content'].apply(summarize_content)
```

```python
df.head()
```

```python
df.to_csv('Finer_aspects_v2_summary.csv', index = False)
```
