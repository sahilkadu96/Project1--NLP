```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
```

```python
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

```python
df = pd.read_csv('Dummy_data_soft_skills.csv')
df
```

```python
def cos_sim(p1, p2):
    
    p1_embed = model.encode(p1)
    p2_embed = model.encode(p2)
    similarity = cosine_similarity([p1_embed], [p2_embed])
    return similarity[0]



def paragraph_similarity_index(dataframe):

    skills = dataframe['Title'].unique()                           #finding out unique skills                                        
    indices = set()                                                  #creating a set of indices

    for i in range(len(skills)):
        print(skills[i])
        content_skill = dataframe[dataframe['Title'] == skills[i]]['Content']           #segregating content as per each skill
        idx = content_skill.index                                                       #noting down the index for each paragraph

        for i in range(idx[0], idx[len(idx)-1]):                                        #we wil only compare adjacent para
                sim = cos_sim(content_skill[i], content_skill[i+1])[0]
                if sim > 0.7:
                  indices.add(i+1)                                                        #appending jth index in list if similarity > 0.7
                  print(f'Similarity between {i} paragraph and {i+1} paragraph is {sim}')
        print("--------------------------")
    return indices  
```

```python
indices  = paragraph_similarity_index(df)
```
